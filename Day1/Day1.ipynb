```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "zepto_v1.xlsx"
```

```python
df = pd.read_excel(file_path)
```

```python
sns.set(style="whitegrid")
```

# Plot 1: MRP vs Discounted Selling Price (Top 10 items)

```python
plt.figure(figsize=(12, 6))
top_items = df.head(10)
x = range(len(top_items))
plt.bar(x, top_items['mrp'], width=0.4, label='MRP', align='center')
plt.bar([i + 0.4 for i in x], top_items['discountedSellingPrice'], width=0.4, label='Discounted Price', align='center')
plt.xticks([i + 0.2 for i in x], top_items['name'], rotation=45, ha='right')
plt.ylabel('Price')
plt.title('MRP vs Discounted Price (Top 10 Items)')
plt.legend()
plt.tight_layout()
plt.show()
```

# Plot 2: Scatter Plot - Discount Percent vs Discounted Selling Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discountPercent', y='discountedSellingPrice', hue='Category', data=df)
plt.title('Discount % vs Discounted Price')
plt.xlabel('Discount Percentage')
plt.ylabel('Discounted Selling Price')
plt.tight_layout()
plt.show()
```

# Plot 3: Histogram - Weight Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['weightInGms'], bins=15, kde=True)
plt.title('Weight Distribution of Items')
plt.xlabel('Weight (gms)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# Plot 4: Category-wise Average Discount %

```python
plt.figure(figsize=(10, 6))
category_avg_discount = df.groupby('Category')['discountPercent'].mean().sort_values()
sns.barplot(x=category_avg_discount.values, y=category_avg_discount.index)
plt.title('Average Discount Percent by Category')
plt.xlabel('Average Discount Percent')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```
